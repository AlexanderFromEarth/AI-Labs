```python
from expert import expertize
from db import process, evidence_names, evidence_values, hypothesis_names, hypothesis_values
from file import load
```

```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import ipywidgets as ws
from IPython.display import display
```

```python
data = process(load("data/db.csv"), ["desease", "ph"], ["symptom", "peh", "penh"])
en = evidence_names(data)
ev = evidence_values(data)
hn = hypothesis_names(data)
hv = hypothesis_values(data)
```

```python
checkboxes = [ws.Checkbox(value=False, description=evidence, disable=False) for evidence in en]
btn = ws.Button(description="Выдать заключение")
out = ws.Output()
```

```python
def predict(btn):
    with out:
        out.clear_output()
        res = dict(zip(hn, expertize(hv, ev, [checkbox.value for checkbox in checkboxes])))
        print(res)
        print(max(res, key=res.get))

btn.on_click(predict)
```

```python
display(*checkboxes, btn, out)
```
